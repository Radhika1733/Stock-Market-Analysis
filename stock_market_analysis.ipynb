```python
# Loading the dataset

stocks_data = pd.read_csv(r"C:\Users\Radhika Sharma\OneDrive\Desktop\stocks.csv")
```

```python
pd.set_option('display.max.rows', 300)
pd.set_option('display.max.columns', 10)
```

```python
stocks_data['Date'] = pd.to_datetime(stocks_data['Date'])

stocks_data['Date'] = stocks_data['Date'].dt.date

print(stocks_data)
```

```python
# Descriptive Statistics for each stock

descriptive_stats = stocks_data.groupby('Ticker')['Close'].describe()
descriptive_stats
```

```python
# Time Series Analysis

pivot_data = stocks_data.pivot(index='Date', columns='Ticker', values='Close')
```

```python
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(8,4))

for column in pivot_data.columns:
    axs.plot(pivot_data.index, pivot_data[column], label=column)

axs.xaxis.set_major_formatter(mdates.DateFormatter('%d %B'))  
axs.xaxis.set_major_locator(mdates.DayLocator(interval=15))
    
axs.legend(loc='upper left', bbox_to_anchor=(1, 1))
axs.set_xlabel('Date')
axs.set_ylabel('Close Price')
axs.set_title('Time Series of Closing Prices')

plt.setp(axs.xaxis.get_majorticklabels(), rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
# Volatility Analysis

volatility = pivot_data.std().sort_values(ascending=False)

fig, axs = plt.subplots(figsize=(8,4))
axs.bar(volatility.index, volatility.values, color='skyblue')

axs.set_xlabel('Ticker')
axs.set_ylabel('Standard Deviation')
axs.set_title('Volatility of Closing Prices (Standard Deviation)')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis

correlation_matrix = pivot_data.corr()

fig, ax = plt.subplots(figsize=(8,4))

cax = ax.imshow(correlation_matrix, cmap='Blues', interpolation='nearest')

ax.set_xticks(np.arange(len(correlation_matrix.columns)))
ax.set_yticks(np.arange(len(correlation_matrix.columns)))
ax.set_xticklabels(correlation_matrix.columns, rotation=45, ha='right')
ax.set_yticklabels(correlation_matrix.columns)

cbar = fig.colorbar(cax)
cbar.set_label('Correlation')

ax.set_title('Correlation Matrix of Closing Prices')

plt.tight_layout()
plt.show()
```

```python
# Comparative Analysis (Calculating the percentage change in closing prices)

percentage_change = ((pivot_data.iloc[-1] - pivot_data.iloc[0]) / pivot_data.iloc[0]) * 100

fig, ax = plt.subplots(figsize=(8,4))

ax.bar(percentage_change.index, percentage_change.values, color='lightblue')

ax.set_ylabel('Percentage Change (%)')
ax.set_xlabel('Ticker')
ax.set_title('Percentage Change in Closing Prices')

plt.xticks(rotation=45) 

plt.tight_layout() 

plt.show()
```

```python
# Daily Risk Vs. Return Analysis

daily_returns = pivot_data.pct_change().dropna()

avg_daily_return = daily_returns.mean()
risk = daily_returns.std()

risk_return_df = pd.DataFrame({'Risk': risk, 'Average Daily Return': avg_daily_return})
```

```python
plt.figure(figsize=(8,4))
plt.scatter(risk_return_df['Risk'], risk_return_df['Average Daily Return'], s=100, color='darkred', alpha=0.7)

for i, txt in enumerate(risk_return_df.index):
    plt.text(risk_return_df['Risk'].iloc[i], risk_return_df['Average Daily Return'].iloc[i], txt,
             horizontalalignment='center', verticalalignment='bottom')

plt.title('Risk vs. Return Analysis')
plt.xlabel('Risk (Standard Deviation)')
plt.ylabel('Average Daily Return')

plt.grid()
plt.tight_layout()  
plt.show()
```
